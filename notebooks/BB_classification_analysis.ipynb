```python
from rdkit.Chem.Draw import MolsToGridImage
from IPython.display import display, Image
import tempfile, os

bbmode_path = '/home/nubar/Downloads/Internship_Chemoinfo/03_synthon/outputs/test_10_BBmode.smi'

mols, legends = [], []
with open(bbmode_path) as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 4:
            mol = Chem.MolFromSmiles(parts[0])
            if mol:
                mols.append(mol)
                legends.append(f"{parts[1]}\n{parts[2]}\n{parts[3]} synthons")

img = MolsToGridImage(mols, molsPerRow=3, subImgSize=(400,300), legends=legends)

# Save using PIL
out_path = '/home/nubar/Downloads/Internship_Chemoinfo/04_results/classified_bbs.png'
img.save(out_path)
display(Image(out_path))
print(f'Showed {len(mols)} classified BBs')
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

print("=== SMALL PRODUCTS (likely simple + simple reactions) ===\n")
with open('/home/nubar/Downloads/Internship_Chemoinfo/03_synthon/outputs/test_extended_enumeration/FinalOut_allEnumeratedCompounds_DuplicatesCanBePresent.smi') as f:
    for line in f:
        smi = line.strip().split()[0]
        mol = Chem.MolFromSmiles(smi)
        if mol:
            mw = Descriptors.MolWt(mol)
            if mw < 250:
                print(f"SMILES: {smi}")
                print(f"MW: {mw:.1f}")
                print("-" * 40)
```
